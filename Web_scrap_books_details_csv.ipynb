```python
def book_data(dat,b1_url):
    title=dat.h1.string
    price=dat.find(class_='price_color').string
    price=float(re.search('[\d.]+',price).group())
    qty=dat.find(class_='instock availability')
    qty=qty.contents[-1].strip()
    qty=int(re.search('\d+',qty).group())
    book_details.append([title,b1_url,price,qty])
    return;
```

```python
curr_url='https://books.toscrape.com/catalogue/page-3.html'
base_url='https://books.toscrape.com/catalogue/'
response=requests.get(curr_url)

k=3;
while(k<=20):   
    data=BeautifulSoup(response.text,'html.parser')
    books=data.find_all(class_='product_pod')
    for i in books:
        b_url=base_url +i.h3.a['href']
        print(b_url)
        respnse=requests.get(b_url)
        dat=BeautifulSoup(respnse.text,'html.parser')
        book_data(dat,b_url)
        
    next_page=data.find(class_='next')
    if(next_page is None):
        break
    next_page_url=base_url + next_page.a['href']
    curr_url=next_page_url
    response=requests.get(curr_url)
    k=k+1
```

```python
len(book_details),curr_url,k
```

```python
df=pd.DataFrame(book_details,columns=['Title','Link','Price','Stock'])
df
```
